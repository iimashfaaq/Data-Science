<h2>Read Data</h2>

Here we will be fetching the data from the above mentioned online source. We will be using `pandas.read_csv()` to read the dataset into a dataframe. The address is put in double quotes.

```python
# Already imported pandas library

# Assign teh address to a variable for easier usage
original_path = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

df = pd.read_csv(original_path, header=None)
```

We have assigned the header to None as this data set does not have any headers present. We will be adding in the header later on.

<h3>Adding Headers</h3>

Pandas has automatically set the column values to intgers staring from 0. We will now add column names provided at <a href="https://archive.ics.uci.edu/ml/datasets/Automobile" target="_blank">https://archive.ics.uci.edu/ml/datasets/Automobile</a>.

```python
# Creating header list

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("Headers: ", headers)
```

Now, we will replace the dataframe column headers with the list we created above. This is done using `dataframe.columns = headers`

```python
df.columns = headers
df.head()
```

We can now see that the header names are set.

Since our target here is to compare prices for different automobiles, we can drop any record that has a missing value of `prices` as follows.

```python
df.dropna(subset=["price"], axis=0)
```

### Identify missing values
#### Convert "?" to NaN

In the car dataset, missing data comes with the question mark "?".
We replace "?" with NaN (Not a Number), which is Python's default missing value marker, for reasons of computational speed and convenience. Here we use the function: 
 <pre>.replace(A, B, inplace = True) </pre>
to replace A by B

```python
import numpy as np

# Replace "?" to NaNs
df.replace("?", np.nan, inplace = True)
df.head()
```

dentify_missing_values

#### Evaluating for Missing Data

The missing values are converted to Python's default. We use Python's built-in functions to identify these missing values. There are two methods to detect missing data:

1. **.isnull()**
2. **.notnull()**

The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

```python
missing_data = df.isnull()
missing_data.head()
```

"True" stands missing values while "False" stands for not missing values

#### Counting missing values in each column

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

### Deal with missing data
**How to deal with missing data?**

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
        c. replace it based on other functions
    </li>
</ol>

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>
<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>
<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four". 
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>
<ul>
    <li>"price": 4 missing data, simply delete the whole row

            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>

#### Calculate average of a column

```python
# Calculating the mean for column "normalized-loss"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("the average normalized-losses: ", avg_norm_loss)
```

#### Raplace NaN with average

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

Similarly, we will be finding out mean and replacing NaN values with the mean value of the corresponding columns.

```python
# Calculating mean for column "bore"
avg_bore = df["bore"].astype("float").mean(axis = 0)
print("The average bore: ", avg_bore)
```

```python
# Replace nan with average in the "bore" column 
df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
# Calculating the mean for column "stroke"
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("The avearge for column stroke is: ", avg_stroke)
```

```python
# Replace NaN values with mean
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
# Calculating mean for column "horsepower"
avg_horse_power = df["horsepower"].astype("float").mean(axis = 0)
print("The mean for column horsepower: ", avg_horse_power)
```

```python
# Replace all NaNs with the mean value
df["horsepower"].replace(np.nan, avg_horse_power, inplace = True)
```

```python
# Calculating mean for column "horsepower"
avg_peak_rpm = df["peak-rpm"].astype("float").mean(axis = 0)
print("The mean for column peak-rpm: ", avg_peak_rpm)
```

```python
# Replace all NaNs with the mean value
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace = True)
```
